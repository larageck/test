Exercise5: Write a Python program to find the median of a given array of numbers using vectorizing

```python
##Aufgabe 4
```

```python
m=[2,3,5,7,8,9,11]

#using for loop

array_len = len(m)

if array_len % 2 == 0:
    median = (m[int(array_len/2)] + m[int(array_len/2 -1)])/2
else:
    median = m[int(array_len/2)]

print(median)
```
